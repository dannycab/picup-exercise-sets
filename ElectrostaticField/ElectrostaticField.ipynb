```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import time
%matplotlib inline

matplotlib.rcParams['xtick.direction'] = 'out'
matplotlib.rcParams['ytick.direction'] = 'out'
```

```python
k = 9e9
q1 = 1e-9
x1 = 0
y1 = 0
extent = 1e-4

spacing = 1e-6
x = np.arange(-extent,extent,spacing)
y = np.arange(-extent,extent,spacing)
```

```python
X,Y = np.meshgrid(x,y)
```

```python
V = k*q1/np.sqrt((X-x1)**2+(Y-y1)**2)
```

```python
plt.figure()
CS = plt.contour(X, Y, V,levels=np.arange(0,10e5,1e5))
plt.title('Energy contours for point charge')
frac=0.2
plt.xlim((-frac*extent,frac*extent));
plt.ylim((-frac*extent,frac*extent));
CB = plt.colorbar(CS, shrink=0.8, extend='both');
```

```python
V = k*q1/np.sqrt((X-x1)**2+(Y-y1)**2)+k*q2/np.sqrt((X-x2)**2+(Y-y2)**2)
```

```python
plt.figure()
CS = plt.contour(X, Y, V,levels=np.arange(-10e5,10e5,1e5))
plt.title('Energy contours for dipole')
frac=0.2
plt.xlim((-frac*extent,frac*extent));
plt.ylim((-frac*extent,frac*extent));
CB = plt.colorbar(CS, shrink=0.8, extend='both');
```

```python
k = 9e9;

extent = 1e-3;
spacing = 1e-6;

x = np.arange(-extent,extent,spacing);
y = np.arange(-extent,extent,spacing);

X,Y = np.meshgrid(x,y);

N = 100;
V = 0;

for i in np.arange(0,N):
    
    qsign = np.random.random_integers(-1,high=1);
    while qsign == 0:
        qsign = np.random.random_integers(-1,high=1);
    
    xsign = np.random.random_integers(-1,high=1);
    while xsign == 0:
        xsign = np.random.random_integers(-1,high=1);
    
    ysign = np.random.random_integers(-1,high=1);
    while ysign == 0:
        ysign = np.random.random_integers(-1,high=1);
    
    qr = qsign*np.random.random()*1e-9;
    xr = xsign*np.random.random()*extent;
    yr = ysign*np.random.random()*extent;
    V += k*qr/np.sqrt((X-xr)**2+(Y-yr)**2)
```

```python
CS = plt.contour(X, Y, V,levels=np.arange(-5e7,5e7,1e5))
#plt.clabel(CS, inline=1, fontsize=10)
plt.title('Energy contours for random charges')
frac=1
plt.xlim((-frac*extent,frac*extent));
plt.ylim((-frac*extent,frac*extent));
CB = plt.colorbar(CS, shrink=0.8, extend='both');
```

```python
im = plt.imshow(V, interpolation='none', cmap=cm.viridis)
CS = plt.contour(X,Y,V,levels=np.arange(-5e7,5e7,1e5))
frac=0.2
plt.xlim((-frac*extent,frac*extent));
plt.ylim((-frac*extent,frac*extent));

# make a colorbar for the contour lines
CB = plt.colorbar(CS, shrink=0.8, extend='both')

# We can still add a colorbar for the image, too.
CBI = plt.colorbar(im, orientation='horizontal', shrink=0.8)
```
