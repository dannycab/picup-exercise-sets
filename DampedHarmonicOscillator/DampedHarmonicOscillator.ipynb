```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
k = 10;
m = 0.01;
c = 0.02;
vi = 0;  # initial velocity (m/s)
xi = 1;  # initial position (m)
dt = 0.00001 # time steps
t_steps = 300000;  # number of time steps
```

```python
time = np.zeros(t_steps)
x = np.zeros(t_steps)
v = np.zeros(t_steps)

time[0] = 0
x[0] = xi
v[0] = vi
```

```python
for i in range(0, t_steps-1):
    time[i+1] = (i+1) * dt
    v[i+1] = v[i] - (k/m) * x[i] * dt - (c/m) * v[i] * dt
    x[i+1] = x[i] + v[i+1] * dt
```

```python
plt.plot(time, v)
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Velocity vs. Time')
```

```python
plt.plot(time, x)
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Position vs. Time')
```

```python
plt.plot(x,v)
plt.xlabel('Position (m)')
plt.ylabel('Velocity (m/s)')
plt.title('Phase Space')
```

```python
U = 0.5*k*x**2
T = 0.5*m*v**2
E = T +U
plt.plot(time,T)
plt.plot(time,U)
plt.plot(time,E)
plt.xlabel('Time (s)')
plt.ylabel('Energy (J)')
plt.title('Energy functions')
plt.legend(['Kinetic','Potential','Mechanical'])
```

```python
ps = np.abs(np.fft.fft(x))**2
time_step = 1/300
freqs = np.fft.fftfreq(x.size,time_step)
idx = np.argsort(freqs)
```

```python
plt.plot(freqs[idx], ps[idx])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power Spectrum')
plt.xlim(-0.1,0.1)
```
